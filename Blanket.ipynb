```python
import pandas as pd
import numpy as np
```

```python
months = ['january', 'february', 'march', 
          'april', 'may', 'june', 'july', 
          'august', 'september', 'october', 
          'november', 'december']
```

```python
random_temps = pd.DataFrame(np.random.randint(0,100,size=(12, 2)), columns=list('ix'))
# df = pd.DataFrame(np.random.randint(0,100,size=(15, 4)), columns=list('ABCD'))
random_temps
```

```python
test_data = [5, 15, 26, 36, 48, 55, 66, 77, 88, 99, 100, -1]
```

```python
test_df = pd.DataFrame({'Temp':test_data, 'Months': months})
test_df
# need to set months as index. `test_df.set_index(months)`
```

```python
def color(val):
    if val >= 100:
        color = 'darkred'
    elif (val >= 90) and (val <= 99):
        color = 'red'
    elif (val >= 80) and (val <= 89):
        color = 'orange'
    elif (val >= 70) and (val <=79):
        color = 'yellow'
    elif (val >= 60) and (val <=69):
        color = 'green'
    elif (val >= 50) and (val <=59):
        color = 'darkgreen'
    elif (val >= 40) and (val <=49):
        color = 'lightblue'
    elif (val >= 30) and (val <=39):
        color = 'blue'
    elif (val >= 20) and (val <=29):
        color = 'mediumpurple'
    elif (val >= 10) and (val <=19):
        color = 'darkorchid'
    elif (val >= 0) and (val <=9):
        color = 'hotpink'
    else:
        color = 'magenta'
    return 'background-color: %s' % color
```

```python
test_df.style.applymap(color, subset=['Temp'])
```

```python
random_temps.style.applymap(color, subset=['i', 'x'])
```
